```python
# Criar uma função de categorização

def categoriza(inv):
    if inv >= 1500:
        return 'Big'
    elif (inv >= 600) and (inv < 1500):
        return 'Mediun'
    elif inv < 600:
        return 'Small'
```

```python
# chamando o método

df['cat_size'] = df['Area'].apply(categoriza)
df.head()
```

```python
# tamanhos categorizados
df.cat_size.value_counts()
```

```python
df.loc[df['Bairro'] == 'Botafogo']
```

## Utilizando Groupby
link: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html

```python
# O mesmo resultado que o loc

df.groupby("Bairro").get_group('Botafogo')
```

```python
df.groupby("cat_size").get_group('Big')[:10]
```

```python
df.groupby('cat_size').last()
```

```python
df.groupby('cat_size').first()
```

```python
# Agrupando por ID

df.groupby('cat_size').groups
```

```python
# Agrupando por size

df.groupby('cat_size').size()
```

```python
# Agrupando categoria, condominio e o valor

df.groupby('cat_size').agg({'Condominio':'count',
                           'Valor':'mean'})
```

```python
# Agrupando Tipo, condominio e o valor

df.groupby('Tipo').agg({'Condominio':'count',
                           'Valor':'mean'})
```

```python
# agrupando um tipo para duas métricas

df.groupby('Tipo').agg(['sum','mean'])
```

```python
df.head()
```

```python
df.groupby('Bairro')[['IPTU','Valor']].sum()
```

```python
# agrupamento po tipo e bairro

df.groupby(by=df['Tipo']).Bairro.value_counts()
```

```python
grouped = df[['Valor','Bairro']].groupby(df['Tipo'])
grouped.mean()
```

```python
df2 = pd.read_csv('data/aluguel2.csv', sep=';')
df2.head()
```

```python
# Valor bruto mensal para um inquilino

df2['Val_mes'] = df2['Valor'] + df2['Condominio'] + df2['IPTU']
df2.head()
```

```python
# Valor do metro quadrado

df2['Val_mq'] = (df2['Valor'] / df2['Area']).round(2)
df2.head()
```

```python
df2.Tipo.value_counts()
```

```python
# agrupando por tipo de residência

df2.groupby('Tipo').get_group('Casa')
```

```python
# Lista com os tipos de casa

casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
df2['Tipo_casa'] = df2['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
df2.head()
```

```python
df2.Tipo_casa.value_counts()
```
